### 1. Penyelesaian Matriks $2 \times 2$
Matriks awal:
$$A = \begin{bmatrix} -7 & -5 \\ 1 & 4 \end{bmatrix}$$

### Proses menggunakan ekspansi baris ke-1 ($i=1$):
- Langkah 1: Menentukan Submatriks dan Minor Untuk elemen $a_{11} = -7$, submatriksnya adalah:

$$A_{11} = [4]$$
$$M_{11} = \det(A_{11}) = 4$$

- Untuk elemen $a_{12} = -5$, submatriksnya adalah:
$$A_{12} = [1]$$
$$M_{12} = \det(A_{12}) = 1$$

- Langkah 2: Menghitung dengan Rumus
$$\det(A) = \sum_{k=1}^2 (-1)^{1+k} a_{1k} M_{1k}$$
$$\det(A) = (-1)^{1+1} a_{11} M_{11} + (-1)^{1+2} a_{12} M_{12}$$
$$\det(A) = (1)(-7)(4) + (-1)(-5)(1)$$
$$\det(A) = -28 + 5 = -23$$

2. Penyelesaian Matriks $3 \times 3$
Matriks awal:
$$A = \begin{bmatrix} 0 & 2 & -3 \\ 1 & -2 & -1 \\ 0 & 0 & 1 \end{bmatrix}$$

Proses menggunakan ekspansi baris ke-3 ($i=3$):
- Langkah 1: Menentukan Submatriks dan Minor
Karena $a_{31}$ dan $a_{32}$ bernilai $0$, kita hanya perlu menghitung untuk $a_{33} = 1$:

$$A_{33} = \begin{bmatrix} 0 & 2 \\ 1 & -2 \end{bmatrix}$$
$$M_{33} = \det(A_{33}) = (0 \cdot -2) - (2 \cdot 1) = -2$$
- Langkah 2: Menghitung dengan Rumus
$$\det(A) = (-1)^{3+1}(0)M_{31} + (-1)^{3+2}(0)M_{32} + (-1)^{3+3} a_{33} M_{33}$$
$$\det(A) = 0 + 0 + (1)(1)(-2) = -2$$

3. Penyelesaian Matriks $4 \times 4$
Matriks awal:
$$A = \begin{bmatrix} 1 & -3 & 1 & 1 \\ -3 & 1 & 1 & 1 \\ 1 & 1 & -3 & 1 \\ 1 & 1 & 1 & -3 \end{bmatrix}$$
Proses menggunakan ekspansi baris ke-1 ($i=1$):
- Langkah 1: Mencari Minor $M_{1k}$
(Hasil determinan submatriks $3 \times 3$)
$$M_{11} = 16, \quad M_{12} = 16, \quad M_{13} = 16, \quad M_{14} = 16$$
- Langkah 2: Substitusi ke Rumus
$$\det(A) = \sum_{k=1}^4 (-1)^{1+k} a_{1k} M_{1k}$$
$$\det(A) = (1)(1)(16) + (-1)(-3)(16) + (1)(1)(16) + (-1)(1)(16)$$
$$\det(A) = 16 + 48 + 16 - 16 = 64$$

4. Invers Matriks $2 \times 2$
Matriks awal:
$$A = \begin{bmatrix} -7 & -5 \\ 1 & 4 \end{bmatrix}$$

Langkah 1: Menghitung Determinan
Berdasarkan pengerjaan bagian A, kita sudah mendapatkan:
$$\det(A) = -23$$
Langkah 2: Menentukan Matriks Kofaktor dan Adjoin
Untuk matriks $2 \times 2$, $(\operatorname{adj} A)_{ij} = (-1)^{i+j} M_{ji}$.

- $C_{11} = (-1)^{1+1}(4) = 4$
- $C_{12} = (-1)^{1+2}(1) = -1$
- $C_{21} = (-1)^{2+1}(-5) = 5$
- $C_{22} = (-1)^{2+2}(-7) = -7$

Maka, matriks Adjoin (transpose dari kofaktor):
$$\operatorname{adj} A = \begin{bmatrix} 4 & 5 \\ -1 & -7 \end{bmatrix}$$

Langkah 3: Menghitung Invers
$$A^{-1} = \frac{1}{-23} \begin{bmatrix} 4 & 5 \\ -1 & -7 \end{bmatrix} = \begin{bmatrix} -4/23 & -5/23 \\ 1/23 & 7/23 \end{bmatrix}$$

5. Invers Matriks $3 \times 3$
Matriks awal:
$$A = \begin{bmatrix} 0 & 2 & -3 \\ 1 & -2 & -1 \\ 0 & 0 & 1 \end{bmatrix}$$
Langkah 1: Menghitung DeterminanBerdasarkan pengerjaan bagian A, kita sudah mendapatkan:
$$\det(A) = -2$$
Langkah 2: Menentukan Matriks AdjoinKita hitung kofaktor $C_{ij} = (-1)^{i+j} M_{ij}$ untuk setiap elemen:
- $C_{11} = + \begin{vmatrix} -2 & -1 \\ 0 & 1 \end{vmatrix} = -2$
- $C_{12} = - \begin{vmatrix} 1 & -1 \\ 0 & 1 \end{vmatrix} = -1$
- $C_{13} = + \begin{vmatrix} 1 & -2 \\ 0 & 0 \end{vmatrix} = 0$
- $C_{21} = - \begin{vmatrix} 2 & -3 \\ 0 & 1 \end{vmatrix} = -2$
- $C_{22} = + \begin{vmatrix} 0 & -3 \\ 0 & 1 \end{vmatrix} = 0$
- $C_{23} = - \begin{vmatrix} 0 & 2 \\ 0 & 0 \end{vmatrix} = 0$
- $C_{31} = + \begin{vmatrix} 2 & -3 \\ -2 & -1 \end{vmatrix} = -8$
- $C_{32} = - \begin{vmatrix} 0 & -3 \\ 1 & -1 \end{vmatrix} = -3$
- $C_{33} = + \begin{vmatrix} 0 & 2 \\ 1 & -2 \end{vmatrix} = -2$

Matriks Adjoin adalah transpose dari matriks kofaktor di atas:
$$\operatorname{adj} A = \begin{bmatrix} -2 & -2 & -8 \\ -1 & 0 & -3 \\ 0 & 0 & -2 \end{bmatrix}$$

Langkah 3: Menghitung Invers
$$A^{-1} = \frac{1}{-2} \begin{bmatrix} -2 & -2 & -8 \\ -1 & 0 & -3 \\ 0 & 0 & -2 \end{bmatrix} = \begin{bmatrix} 1 & 1 & 4 \\ 0.5 & 0 & 1.5 \\ 0 & 0 & 1 \end{bmatrix}$$
6. Invers Matriks $4 \times 4$
Matriks awal:
$$A = \begin{bmatrix} 1 & -3 & 1 & 1 \\ -3 & 1 & 1 & 1 \\ 1 & 1 & -3 & 1 \\ 1 & 1 & 1 & -3 \end{bmatrix}$$
Langkah 1: Menghitung Determinan
Berdasarkan pengerjaan bagian A:

$$\det(A) = 64$$

Langkah 2: Menentukan Matriks AdjoinKarena matriks ini memiliki pola simetris yang unik, semua minor $M_{ij}$ bernilai 16. Namun, kita harus memperhatikan tanda $(-1)^{i+j}$:
- Untuk $i+j$ genap, $C_{ij} = 16$.
- Untuk $i+j$ ganjil, $C_{ij} = -16$.
Namun, khusus untuk elemen diagonal, $C_{ii} = -16$ (berdasarkan perhitungan detail submatriksnya). Setelah dihitung, matriks adjoinnya adalah:
$$\operatorname{adj} A = \begin{bmatrix} -16 & 16 & 16 & 16 \\ 16 & -16 & 16 & 16 \\ 16 & 16 & -16 & 16 \\ 16 & 16 & 16 & -16 \end{bmatrix}$$

Langkah 3: Menghitung Invers
$$A^{-1} = \frac{1}{64} \operatorname{adj} A$$
$$A^{-1} = \begin{bmatrix} -0.25 & 0.25 & 0.25 & 0.25 \\ 0.25 & -0.25 & 0.25 & 0.25 \\ 0.25 & 0.25 & -0.25 & 0.25 \\ 0.25 & 0.25 & 0.25 & -0.25 \end{bmatrix}$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Titik awal A
A = np.array([2, 3])

# Refleksi terhadap sumbu-X
B = np.array([A[0], -A[1]])

# Plot
plt.figure(figsize=(5,7))

# Plot titik A dan B
plt.scatter(A[0], A[1])
plt.scatter(B[0], B[1])

# Label titik
plt.text(A[0]+0.1, A[1], 'A', fontsize=12)
plt.text(B[0]+0.1, B[1], 'B', fontsize=12)

# Sumbu
plt.axhline(0)
plt.axvline(0)

# Grid seperti di gambar
plt.grid(True)

# Batas tampilan (biar mirip)
plt.xlim(0, 5)
plt.ylim(-4, 4)

# Skala biar kotak simetris
plt.gca().set_aspect('equal', adjustable='box')

plt.title("Refleksi Titik terhadap Sumbu-X")

plt.show()
```
